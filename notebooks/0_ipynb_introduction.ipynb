# What is a Notebook?

A notebook file, the one ending with `.ipynb` - which stands for "*IPython Notebook*" - is very usefull during Data Analysis and Software Development since it allows to put code lines and annotations togheter.

A notebook has some caracteristics:
- has `code` cell that allow to write and run code,
- has `markdown` cell that allow to write [MarkDown](https://www.markdownguide.org/basic-syntax/) text annotations to explain and integrate your code.
- in Markdown cells it allows to embedd tables, images and hyperlinks, so that you can have a reference to an external source.

# MarkDown cells

+ To ***add a New Markdown cell Below*** the cell you are actually running, ***press*** ***B*** (= add bottom), then ***M*** (= markdown) 
+ To ***add a New Markdown cell Above*** the cell you are actually running, ***press*** ***A*** (= add above), then ***M*** (= markdown)  
+ To ***delete a Markdown cell***, select the cell and, ***press*** ***twice D*** = DD)  



**To write** inside a Markdown cell, select it and press *ENTER*.\
**After writing**, to make it effective press *CTRL + ENTER*

### Headings (#)

With markdown cells you can create differrent **chapters** in your document, to give them a hierarchy you can insert an `hashtag` (#) before a heading title to give the title an heading level 1.

You can add hasthags up to level 6 as shown below:

___

___

___

### Heading level 3
___

#### Heading level 4
___

\##### Heading level 5 (write this to render the appearence below)
##### Heading level 5
___

\###### Heading level 6 (write this to render the appearence below)
##### Heading level 6
___

### **Bold**, *Italics* and ***Bold-italics*** text

- To write *italic* text, include the text between single asteriks (i.e., \*text\*) - or select the text and press CTRL + I if using LINUX
- To write **bold** text, include the text between double asteriks (i.e., \*\*text\*\*)  - or select the text and press CTRL + B if using LINUX
- To write ***italic-bold*** text, include the text between triple asteriks (i.e., \*\*\*text\*\*\*) - or select the text and press CTRL + I, then CTRL + B if using LINUX
