# Image Data and Analysis in Python

In this notebook, we will explore how to handle and analyze image data using Python. We will use libraries like [`OpenCV`](https://opencv.org/) and [`Matplotlib`](https://matplotlib.org/) to load, display, and manipulate images. This will help us understand the basic concepts of image processing.


import libraries

```python
import cv2
import matplotlib.pyplot as plt
```

## Loading and Displaying Images

To start, let's see how to load an image from a file and display it. We will use the `OpenCV` library for loading and `Matplotlib` for displaying the image.

`OpenCV` imports images as default option in *Blue-Green-Red* colorspace, while we are used to watch digital images in *Red-Green-Blue* colorspace. A colorspace conversion is thus needed.

```python
# Load an image
image_path = '../data/images/fruits.jpg'

# open image in BGR colorspace
image_bgr = cv2.imread(image_path)

# Convert the image from BGR (OpenCV default) to RGB (Matplotlib default)
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis('off')  # Hide axes
plt.title("Image")
plt.show()
```

## Get image info

```python
# Load an image
image_path = '../data/images/fruits.jpg'

# open image in BGR colorspace
image_bgr = cv2.imread(image_path)

print(f'image dimension (h, w, c): {image_rgb.shape}')
print(f'number of pixels: {image_rgb.size}')
print(f'get data type: {image_rgb.dtype}')
print(f'get memory usage [Bytes]: {image_rgb.nbytes}')
```

## Decomposing and image into its channels

**`BGR`** and **`RGB`** images are multi-channels arrays (3 channels), while **grayscale** images have a single-channel. Here below how to split an image into its components.

```python
# Load an image
image_path = '../data/images/fruits.jpg'

# open image in BGR colorspace
image_bgr = cv2.imread(image_path)

# Convert the image from BGR (OpenCV default) to RGB (Matplotlib default)
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# decompose the image into red, green and blue
red_ch, g_ch, blue_ch = cv2.split(image_rgb)

# Display the image
plt.imshow(red_ch, cmap='Reds')
plt.axis('off')  # Hide axes
plt.title("Red channel")
plt.show()

plt.imshow(g_ch, cmap='Greens')
plt.axis('off')  # Hide axes
plt.title("Green channel")
plt.show()

plt.imshow(blue_ch, cmap='Blues')
plt.axis('off')  # Hide axes
plt.title("Blue channel")
plt.show()
```

## Converting Images to Grayscale

Often, it's useful to convert images to grayscale to simplify analysis. Let's see how to do this using `OpenCV`.

```python
# Load an image
image_path = '../data/images/fruits.jpg'
image_bgr = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image")
plt.show()
```

## Image pixel values distribution (histogram)

An image histogram shows the distribution of pixel values. This can be useful to understand the contrast and brightness of the image.

```python
# Load an image
image_path = '../data/images/fruits.jpg'
image_bgr = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

# Calculate the histogram
histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

# Plot the histogram
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Pixel value")
plt.ylabel("Frequency")
plt.plot(histogram)
plt.xlim([0, 256])
plt.show()
```

## Edge detection

Edge detection is a fundamental operation in image processing. We will use the Canny edge detection algorithm provided by `OpenCV`.

```python
# Load an image
image_path = '../data/images/fruits.jpg'
image_bgr = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

# Apply Canny edge detection
edges = cv2.Canny(gray_image, threshold1=100, threshold2=200)

# Display the edges
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.title("Canny Edge Detection")
plt.show()
```

## Clip an image

```python
# Load an image
image_path = '../data/images/fruits.jpg'
image_bgr = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

# image clipping [h1 : h2 , w1 : w2]
clipped_image = gray_image[10 : 120, 30 : 190]

# Display the edges
plt.imshow(clipped_image, cmap='gray')
plt.axis('off')
plt.title("Clipped image")
plt.show()
```
